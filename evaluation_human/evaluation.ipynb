```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
agents = ['RAG', '4o', '4o_mini', '3.5']
correct = [27, 20, 14, 0]
incorrect = [0, 5, 6, 0]
no_answer = [3, 5, 10, 30]
percent_correct = [90, 66.66, 46.66, 0]

# Create the stacked bar chart
fig, ax1 = plt.subplots(figsize=(8, 6))  # Increased figure size

bar_width = 0.4
index = np.arange(len(agents))

p1 = ax1.bar(index, correct, bar_width, label='Correct', color='g')
p2 = ax1.bar(index, incorrect, bar_width, bottom=correct, label='Incorrect', color='r')
p3 = ax1.bar(index, no_answer, bar_width, bottom=np.array(correct) + np.array(incorrect), label='No Answer', color='gray')

ax1.set_xlabel('Agents', fontsize=14)
ax1.set_ylabel('Number of Responses', fontsize=14)
ax1.set_title('Performance Comparison of RAG vs Other LLM Agents', fontsize=16)
ax1.set_xticks(index)
ax1.set_xticklabels(agents, rotation=0, fontsize=12)  # Removed rotation

# Increase tick label font size
ax1.tick_params(axis='both', which='major', labelsize=12)

# Add the line chart for percentage correct
ax2 = ax1.twinx()
line = ax2.plot(index, percent_correct, color='blue', marker='o', linestyle='-', linewidth=2, markersize=8, label='% Correct')
ax2.set_ylabel('Percentage Correct', fontsize=14)
ax2.set_ylim(0, 100)
ax2.tick_params(axis='y', labelsize=12)

# Add percentage labels on the line
for i, v in enumerate(percent_correct):
    ax2.text(i, v + 2, f'{v}%', ha='center', va='bottom', color='blue', fontsize=12)

# Combine legends
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()

# Place legend below the plot
plt.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4, fontsize=12)

plt.tight_layout()
plt.subplots_adjust(bottom=0.2)  # Adjusted bottom margin
plt.rcParams['font.family'] = 'Arial'  # Use a clear font

plt.show()
```
